# 03 Model Architecture

Inspect the Generator and Discriminator shapes used by the project.

```python
import sys
from pathlib import Path

import torch

REPO_ROOT = Path('..').resolve()
sys.path.append(str(REPO_ROOT))

from gan_anime_faces import Discriminator, Generator
```

```python
generator = Generator(latent_dim=100)
discriminator = Discriminator()

noise = torch.randn(4, 100, 1, 1)
fake_images = generator(noise)
decision = discriminator(fake_images)

print('Generator output shape:', fake_images.shape)
print('Discriminator output shape:', decision.shape)
print('Generator parameters:', sum(p.numel() for p in generator.parameters()))
print('Discriminator parameters:', sum(p.numel() for p in discriminator.parameters()))
```
