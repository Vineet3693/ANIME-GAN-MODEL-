# 06 Evaluation

Run FID evaluation after training and inspect the saved metrics.

```python
import json
import subprocess
import sys
from pathlib import Path

command = [
    sys.executable,
    '../gan_anime_faces.py',
    'evaluate',
    '--data-dir', '/kaggle/input/animefacedataset',
    '--checkpoint', '../outputs/checkpoints/latest.pt',
    '--output-dir', '../outputs',
    '--eval-samples', '5000',
]
subprocess.run(command, check=True)

metrics_path = Path('../outputs/reports/evaluation_metrics.json')
json.loads(metrics_path.read_text(encoding='utf-8'))
```
