# 01 Data Exploration

Use this notebook on Kaggle to inspect dataset size, image formats, and sample images before training.

```python
from collections import Counter
from pathlib import Path

import matplotlib.pyplot as plt
from PIL import Image

DATA_DIR = Path('/kaggle/input/animefacedataset')
IMAGE_EXTENSIONS = {'.jpg', '.jpeg', '.png', '.bmp', '.webp'}
image_paths = sorted([path for path in DATA_DIR.rglob('*') if path.suffix.lower() in IMAGE_EXTENSIONS])
print(f'Total images: {len(image_paths)}')
print('Format counts:', Counter(path.suffix.lower() for path in image_paths).most_common())
```

```python
sample_paths = image_paths[:25]
fig, axes = plt.subplots(5, 5, figsize=(10, 10))
for axis, image_path in zip(axes.flat, sample_paths):
    image = Image.open(image_path).convert('RGB')
    axis.imshow(image)
    axis.set_title(image_path.name[:12], fontsize=8)
    axis.axis('off')
plt.tight_layout()
plt.show()
```
