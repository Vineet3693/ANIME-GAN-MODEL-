# 05 Visualization

Plot loss curves and inspect generated samples saved during training.

```python
import json
from pathlib import Path

import matplotlib.pyplot as plt
from PIL import Image

reports_dir = Path('../outputs/reports')
samples_dir = Path('../outputs/samples')
metrics = json.loads((reports_dir / 'training_metrics.json').read_text(encoding='utf-8'))
history = metrics['history']
epochs = [entry['epoch'] for entry in history]
d_loss = [entry['discriminator_loss'] for entry in history]
g_loss = [entry['generator_loss'] for entry in history]
```

```python
plt.figure(figsize=(8, 5))
plt.plot(epochs, d_loss, label='Discriminator Loss')
plt.plot(epochs, g_loss, label='Generator Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training Loss')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

latest_grid = sorted(samples_dir.glob('epoch_*_grid.png'))[-1]
Image.open(latest_grid)
```
